```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url='https://www.zomato.com/bangalore/restaurants'
```

```python
browser=webdriver.Chrome()
browser.get(url)
```

```python
html=browser.page_source
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
rest=soup.find_all('div',{'class':'sc-1q7bklc-1 cILgox'})
rest[10].text
```

```python
restaurant_name=[]
ratings=[]
price_for_one=[]
cuisins=[]
link=[]

for i in soup.find_all('h4',{'class':'sc-1hp8d8a-0 sc-imDdex fdjWVR'}):
    restaurant_name.append(i.text)
for j in soup.find_all('div',{'class':'sc-1q7bklc-1 cILgox'}):
    ratings.append(j.text)
for k in soup.find_all('p',{'class':'sc-1hez2tp-0 sc-lffWgi liHDRm'}):
    price_for_one.append(k.text)
for m in soup.find_all('p',{'class':'sc-1hez2tp-0 sc-lffWgi hyoGXE'}):
    cuisins.append(m.text)
for l in soup.find_all('a', href = True):
    if 'order' in l['href']:
        if l['href'] not in link:
            
            link.append(l['href'])
```

```python
delivery_review_no=[]
locations=[]
count = 0
for i in link:
    count+=1
    print(count,end=" ")
    r=browser.get('https://www.zomato.com/'+i)
    sou = BeautifulSoup(browser.page_source, 'lxml')
    try:
        j = sou.find_all("a",{"class" : "sc-clNaTc vNCcy"})
        locations.append(j[0].text)
    except:
        locations.append('NA')
        
    try:
        j = sou.find_all("div",{"class" : "sc-1q7bklc-8 kEgyiI"})
        delivery_review_no.append(j[0].text)
    except:
        delivery_review_no.append('NA')
```

```python
max_length = max(len(restaurant_name), len(ratings), len(price_for_one), len(cuisins), len(link), len(locations), len(delivery_review_no))

restaurant_name += [None] * (max_length - len(restaurant_name))
ratings += [None] * (max_length - len(ratings))
price_for_one += [None] * (max_length - len(price_for_one))
cuisins += [None] * (max_length - len(cuisins))
link += [None] * (max_length - len(link))
locations += [None] * (max_length - len(locations))
delivery_review_no += [None] * (max_length - len(delivery_review_no))

data = {'restaurant_name':restaurant_name,'ratings':ratings,'price_for_one':price_for_one,'cuisins':cuisins,'links':link,
    'locations': locations,'delivery_review_no':delivery_review_no}
df = pd.DataFrame(data)


print(df)
```
